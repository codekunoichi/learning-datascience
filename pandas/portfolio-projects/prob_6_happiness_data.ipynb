2. Try to answer the following questions

- What countries or regions rank the highest in overall happiness and each of the six factors contributing to happiness?

Steps
1. Load the data for 2015 till 2019
2. Clean the data and concat all years
3. Identify top twenty country's by each six factors
4. Plot each country year by year

```python
import pandas as pd 

data_2015 = pd.read_csv('../data-files/happiness_data/2015.csv')
data_2016 = pd.read_csv('../data-files/happiness_data/2016.csv')
data_2017 = pd.read_csv('../data-files/happiness_data/2017.csv')
data_2018 = pd.read_csv('../data-files/happiness_data/2018.csv')
data_2019 = pd.read_csv('../data-files/happiness_data/2019.csv')
```

- Observing the data set and the changes in column names, first we will add a column for year to each of the data set.
- Then we will normalize the data set to have the same column name across each year's data set before we will concatenate them.
- Then we will trend the year over year statistics

```python
# Let us add the year to each of the dataframe for the year.
data_2015['year'] = 2015
data_2016['year'] = 2016
data_2017['year'] = 2017
data_2018['year'] = 2018
data_2019['year'] = 2019
data_2015.head()
```

### Normalize columns
- Normalize the column names for each of the similar looking column names across the years
- Rename for each year the columns to the same name. 
- Keep the following for analysis

1. Country
2. Region
3. Happiness Rank
4. Happiness Score
5. Standard Error
6. GDP per capita
7. Family
8. Health Life Expectancy
9. Freedom
10. Trust Govt Corruption
11. Generosity

#### Order them all as follows:
1. Happiness Rank	
2. Country or region	
3. Happiness Score	
4. GDP per Capita	
5. Family	
6. Health Life Expectancy	
7. Freedom	
8. Generosity	
9. Trust Govt Corruption	
10. year


```python
# Define the desired column order
desired_order = ['Happiness Rank', 'Country', 'Happiness Score', 'GDP per Capita', 'Family', 'Health Life Expectancy', 'Freedom', 'Generosity', 'Trust Govt Corruption', 'year']
```

```python
data_2015 = data_2015.rename(columns={'Economy (GDP per Capita)': 'GDP per Capita', 'Health (Life Expectancy)': 'Health Life Expectancy', 'Trust (Government Corruption)': 'Trust Govt Corruption'})
data_2015 = data_2015.drop(['Region', 'Standard Error', 'Dystopia Residual'], axis=1)
data_2015 = data_2015[desired_order]
data_2015.head()
```

```python
data_2016 = data_2016.rename(columns={'Economy (GDP per Capita)': 'GDP per Capita', 'Health (Life Expectancy)': 'Health Life Expectancy', 'Trust (Government Corruption)': 'Trust Govt Corruption'})
data_2016 = data_2016.drop(['Region', 'Lower Confidence Interval', 'Upper Confidence Interval', 'Dystopia Residual'], axis=1)
data_2016 = data_2016[desired_order]
data_2016.head()
```

```python
data_2017 = data_2017.rename(columns={'Happiness.Rank': 'Happiness Rank', 'Happiness.Score' : 'Happiness Score', 'Economy..GDP.per.Capita.': 'GDP per Capita', 'Health..Life.Expectancy.': 'Health Life Expectancy', 'Trust..Government.Corruption.': 'Trust Govt Corruption', 'Dystopia.Residual' : 'Dystopia Residual'})
data_2017 = data_2017.drop(['Whisker.high', 'Whisker.low', 'Dystopia Residual'], axis=1)
data_2017 = data_2017[desired_order]
data_2017.head()
```

```python
data_2018 = data_2018.rename(columns={'Country or region': 'Country', 'Overall rank': 'Happiness Rank', 'Score' : 'Happiness Score', 'GDP per capita': 'GDP per Capita', 'Healthy life expectancy': 'Health Life Expectancy', 'Perceptions of corruption': 'Trust Govt Corruption', 'Freedom to make life choices' : 'Freedom', 'Social support':'Family'})
data_2018 = data_2018[desired_order]
data_2018.head()
```

```python
data_2019 = data_2019.rename(columns={'Country or region': 'Country', 'Overall rank': 'Happiness Rank', 'Score' : 'Happiness Score', 'GDP per capita': 'GDP per Capita', 'Healthy life expectancy': 'Health Life Expectancy', 'Perceptions of corruption': 'Trust Govt Corruption', 'Freedom to make life choices' : 'Freedom', 'Social support':'Family'})
data_2019 = data_2019[desired_order]
data_2019.head()
```

#### Concat all into one dataframe

```python
# List of DataFrames for each year
data_frames = [data_2015, data_2016, data_2017, data_2018, data_2019]  

# Merge the DataFrames iteratively
merged_df = data_frames[0]  # Start with the first DataFrame
for i in range(1, len(data_frames)):
    print(i)
    print(str(2015 + i - 1))
    merged_df = pd.merge(merged_df, data_frames[i], on='Country', how='outer', suffixes=('_' + str(2015 + i - 1), '_' + str(2015 + i)))

merged_df.head()
```

```python
# Get all column headings
column_headings = merged_df.columns.tolist()

# Display the column headings
print(column_headings)
```

#### Sort the merged DF by 2019 Happiness Rank and lets study the top 30 nations

```python
sorted_df = merged_df.sort_values(['Happiness Rank'])
sorted_df.head()
```

```python
top_20 = sorted_df[:20]
top_20.head()
```

#### Lets visualize the top twenty Happy Nations and year over year change

```python
import pandas as pd
pd.plotting.register_matplotlib_converters()
import matplotlib.pyplot as plt
%matplotlib inline
import seaborn as sns
print("Setup Complete")
```

```python
concat_df = pd.concat(data_frames)
concat_df.head()
```

```python
top_20_cntry_data = concat_df.loc[concat_df.Country.isin(data_2019.sort_values('Happiness Rank')[:20]['Country'])]
top_20_cntry_data.head()
#reviews.loc[reviews.country.isin(['Italy', 'France'])]
```

```python
# Set the width and height of the figure
plt.figure(figsize=(14,6))

# Bar Chart showing average arrival delay for Spirit Airlines airline code = NK flights by month
ax = sns.barplot(x=top_20_cntry_data.Country, y=top_20_cntry_data['Happiness Score'])
# Rotate x-labels vertically
ax.set_xticklabels(ax.get_xticklabels(), rotation=90)

# Add label for vertical axis
plt.ylabel("Happiness Score")
```

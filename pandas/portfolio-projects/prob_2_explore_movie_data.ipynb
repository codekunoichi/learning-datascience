### Load dataset into a Pandas DataFrame

````python
import pandas as pd

# Filepath to movie data
movie_file = '../data-files/imdb_top_1000.csv'

## Load CSV into data frame
movie_data = pd.read_csv(movie_file)

movie_data.describe()
````

````python
movie_data.head()
````

### Find the number of movies released each year

````python
movie_data.groupby(['Released_Year']).value_counts()
````

### Determine the most common genres and plot a bar chart to visualize the distribution.

````python
# Lets split the Genre column to aggregate them better

movie_data['Genre'] = movie_data['Genre'].str.split(', ')
movie_data.head()
````

````python
# Step 2: Count the occurrences of each genre
# Using explode to create separate rows for each genre
genre_counts = movie_data.explode('Genre')['Genre'].value_counts()
genre_counts
````

````python
import matplotlib.pyplot as plt

# Step 3: Visualize the distribution of genres
plt.figure(figsize=(8, 6))
genre_counts.plot(kind='bar')
plt.title('Genre Distribution (using only Matplotlib library)')
plt.xlabel('Genre')
plt.ylabel('Count')
plt.show()
````

````python
import seaborn as sns
plt.figure(figsize=(16, 8))
plt.title('Genre Distribution (using only Seaborn library - observe the colors)')
sns.set_theme(style="whitegrid")
new_df = genre_counts.reset_index(name='genre_count')
new_df
# Bar Chart showing average arrival delay for Spirit Airlines airline code = NK flights by month
ax = sns.barplot(x=new_df.Genre, y=new_df.genre_count)
# Rotate x-labels vertically
ax.set_xticklabels(ax.get_xticklabels(), rotation=90)
plt.xlabel("Genre")
plt.ylabel("Genre Counts")
````

### Superimposing two bar graphs together, representing two decades, 1980-1999 and 2000 onwards

````python
import pandas as pd
import matplotlib.pyplot as plt

# Create separate DataFrames for the 2000s and 1980s-1990s decades
movie_data_2000s = movie_data.loc[(movie_data.Released_Year >= 2000)]
movie_data_80s_90s = movie_data.loc[(movie_data.Released_Year >= 1980) & (movie_data.Released_Year < 2000)]

# Group the data by decades and calculate the number of movies released in each decade
decades_2000s = movie_data_2000s.groupby((movie_data_2000s['Released_Year'] // 10) * 10).size()
decades_80s_90s = movie_data_80s_90s.groupby((movie_data_80s_90s['Released_Year'] // 10) * 10).size()

# Superimpose bar graphs for the two decades
plt.figure(figsize=(8, 6))
decades_2000s.plot(kind='bar', position=0, width=0.4, label='2000s', color='blue', alpha=0.7)
decades_80s_90s.plot(kind='bar', position=1, width=0.4, label='80s-90s', color='orange', alpha=0.7)
plt.title('Movie Counts by Decades')
plt.xlabel('Decade')
plt.ylabel('Number of Movies')
plt.legend()
plt.show()
````

### Seaborn and Matplotlib coexisting together, how are they working together

Seaborn (often imported as `sns`) is a high-level Python data visualization library built on top of Matplotlib, which is a powerful 2D plotting library. Seaborn enhances Matplotlib by providing a high-level interface for creating attractive and informative statistical graphics.

When Seaborn and Matplotlib are used together, Seaborn takes advantage of Matplotlib's underlying functionality to create visualizations. Seaborn simplifies the process of creating complex statistical visualizations by providing default themes, color palettes, and convenient functions for common plot types. It also integrates seamlessly with Pandas DataFrames.

Here's how Seaborn and Matplotlib interact:

1. **Seaborn Themes:** Seaborn sets a default theme for Matplotlib plots, making them visually appealing without requiring manual adjustments. You can set a theme using `sns.set_theme()`.

2. **Color Palettes:** Seaborn provides a variety of color palettes that you can use in your Matplotlib plots. These palettes make it easy to distinguish different elements in your visualizations.

3. **Convenient Plot Functions:** Seaborn offers higher-level functions for creating complex plot types, such as bar plots, scatter plots, box plots, and more. These functions often take Pandas DataFrames as input and automatically handle common data visualization tasks.

4. **Seaborn Settings:** You can adjust Seaborn settings using `sns.set()` to control elements like font sizes, background colors, and grid styles, which can influence the appearance of Matplotlib plots.

Despite the close integration, it's important to note that you can still use pure Matplotlib functions alongside Seaborn. For instance, you can combine Seaborn's color palettes with Matplotlib's customizations or use Seaborn for certain plot types and switch to Matplotlib for more specialized customization.

Here's an example of how Seaborn and Matplotlib work together:

```python
import seaborn as sns
import matplotlib.pyplot as plt

# Example data
tips = sns.load_dataset("tips")

# Seaborn plot with a default theme and color palette
sns.barplot(x="day", y="total_bill", data=tips, palette="Set2")

# Matplotlib customization
plt.title("Total Bill by Day (Seaborn + Matplotlib)")
plt.xlabel("Day")
plt.ylabel("Total Bill")

# Show the plot
plt.show()
```

In this example, Seaborn is used to create the initial bar plot with a specified color palette, and Matplotlib is used for additional customizations such as the plot title and axis labels. This illustrates how Seaborn and Matplotlib can work together harmoniously to create informative and visually appealing data visualizations.

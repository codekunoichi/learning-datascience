## Bar Charts and Heatmaps

```python
# path of file to read
flight_delays_filepath = './data-files/flight_delays.csv'

# load CSV with month as index
flight_delay_data = pd.read_csv(flight_delays_filepath, index_col='Month')

# Examine the data
flight_delay_data
```

## Scatter Plots

```python
# Path of the file to read
insurance_filepath = './data-files/insurance.csv'

# read the file into a variable insurance_data
insurance_data = pd.read_csv(insurance_filepath)

# Explore the data set
insurance_data.head()
```

### Simple Scatter Plot to plot BMI to charges

It demonstrates, higher BMI individuals pay for more insurance charges. As higher BMI is typically associated higher risk of chronic conditions.

```python
sns.scatterplot(x=insurance_data['bmi'], y=insurance_data['charges'])
```

```python
# To double-check the strength of this relationship, you might like to add a regression line, or the line that best fits the data.
sns.regplot(x=insurance_data['bmi'], y=insurance_data['charges'])
```

### Color Coded Scatter Plots

To understand how smoking affects the relationship between BMI and insurance costs, we can color-code the points by `'smoker'`, and plot the other two columns (`'bmi', 'charges'`) on the axes

```python
sns.scatterplot(x=insurance_data['bmi'], y=insurance_data['charges'], hue=insurance_data['smoker'])
```

Use sns.lmplot command to add two regression lines, corresponding to smokers and nonsmokers. (You'll notice that the regression line for smokers has a much steeper slope, relative to the line for nonsmokers!)

```python
sns.lmplot(x="bmi", y="charges", hue="smoker", data=insurance_data)
```

 we use scatter plots to highlight the relationship between two continuous variables (like "bmi" and "charges"). However, we can adapt the design of the scatter plot to feature a categorical variable (like "smoker") on one of the main axes. We'll refer to this plot type as a categorical scatter plot, and we build it with the sns.swarmplot command.

```python
sns.swarmplot(x=insurance_data['smoker'], y=insurance_data['charges'])
```

### Create a histogram plot for four different columns of the Iris dataset in a single plot using Seaborn.

```python
import seaborn as sns
import matplotlib.pyplot as plt

# Load the Iris dataset
data = sns.load_dataset('iris')

# Set up the figure and axes
fig, axes = plt.subplots(nrows=1, ncols=4, figsize=(12, 4))

# Create histograms for each column
sns.histplot(data['sepal_length'], kde=True, ax=axes[0], color='blue')
sns.histplot(data['sepal_width'], kde=True, ax=axes[1], color='orange')
sns.histplot(data['petal_length'], kde=True, ax=axes[2], color='green')
sns.histplot(data['petal_width'], kde=True, ax=axes[3], color='red')

# Add titles to subplots
axes[0].set_title('Sepal Length')
axes[1].set_title('Sepal Width')
axes[2].set_title('Petal Length')
axes[3].set_title('Petal Width')

# Adjust layout
plt.tight_layout()

# Show the plot
plt.show()
```

### Explaining the important code line `sns.histplot(data['sepal_length'], kde=True, ax=axes[0], color='blue')`

1. `sns.histplot`: This is a function from the Seaborn library used to create a histogram plot.

2. `data['sepal_length']`: This is the specific column (feature) of the dataset that you want to create a histogram for. In this case, it's the "sepal_length" column from the Iris dataset.

3. `kde=True`: This parameter determines whether to include a Kernel Density Estimate (KDE) plot on top of the histogram. When `kde=True`, it adds a smooth line that estimates the underlying distribution of the data.

4. `ax=axes[0]`: This parameter specifies the subplot (axes) where you want to create the histogram. In this case, it's the first subplot in the `axes` array that was defined earlier.

5. `color='blue'`: This parameter allows you to set the color of the histogram bars. In this case, it sets the color to blue.

So, the entire line of code `sns.histplot(data['sepal_length'], kde=True, ax=axes[0], color='blue')` is creating a histogram plot for the "sepal_length" column of the Iris dataset. It's adding both histogram bars and a KDE curve on top of the bars, with a blue color, and it's placing this plot in the first subplot (axes) of the layout you defined earlier.

You can use similar code lines to create histogram plots for the other columns in the Iris dataset, adjusting the column name, color, and subplot index as needed.
